```
import io
def calculate_bleurt(filename):
  with io.open(filename,'r') as f:
    bl = f.readlines()

  b = [float(i) for i in bl]
  score = sum(b)/len(b)
  print('score for ' + filename)
  print(score)


```

```
calculate_bleurt('bleurt_scores_t5-base-fine-tuned-qg-hl_t5_tweet_manual.txt')
calculate_bleurt('bleurt_scores_t5-small-fine-tuned-qg-hl_t5_tweet_manual.txt')
calculate_bleurt('bleurt_scores__valhalla_base_base-qg-hl_tweet_manual_model.txt')
calculate_bleurt('bleurt_scores_t5-valhalla-small-qg-hl_tweet_manual_model.txt')
```
